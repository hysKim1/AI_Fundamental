##### 파레토 법칙(Pareto Principle)
- 대부분의 판매는 상위 20%가 전체 매출의 80%를 차지한다(전통적)
    - 대부분의 80%가 경시하는 경향이 있음
- 최근 이것이 깨짐

##### 롱태일의 법칙(Long Tail theory)
- 나머지 80%의 중요성을 강조하며 인터넷과 물류가 발단되면서 시작.
- 예. Amazon - 베스트 셀러 외에 나머지 물건들도 돈이 됨
    - 추천 시스템 도입(대체재,보안재) 
    
##### 문제점: 필터 버블(Filter Bubble)
- 사용자 정보 검색물 바탕으로 특정 정보가 사람들의 Bias를 강화시킴
    - Facebook 실험 
        - 한 그룹은 우울한 피드 노출시켰더니 영향을 미침.
        - 주어진 정보가 사람들의 감정에 영향을 준다는 결과
        - 책추천 :생각 조종자들 

##### 1984(오웰)vs 멋진 신세계 (헉슬리)
 - 오웰 : 정부의 통제
 - 헉슬리 : 통제가 없어 정보의 바다에서 진실이 전달 되지 않음

- 높은 유사성 갖은 유저의 높은평점 받은 영화 추천
1. 유저 아이디 입력받으면 그분에게 가장 높은 상관성 갖는사람들 뽑음
2. 뽑고나서 그 k번쨰사람들과의 차집합중에서 평균 평점 높은거 or 곂치는게 많은거



```python
def user_movies(user_id, k):
    return data[data['user id']==user_id]['item id'].values[1:k]
```

데이터프레임에 해당 값이 속해있는지 boolean 값으로 반환 (boolean indexing 방법 이용하는것!)

5> 평점별 가중치 적용
- 유저 123의 상관성 분석에서 자신의 가중치(상관계수)를 넣을 예정

```python
relation.loc[123].sort_values(ascending=False)[1:4]
```

* 각 user id에 따라서 각 조선식을 주어 해당 유저인경우 해당 값을 가중치(평점에 곱하여) 할당
    * 유사도 높을수록 가중치가 높아 영향이 더 큼
* 예. temp에 weight이라는 열을 생성해서 유저 6인경우 0.414398,
* 유저 6아니고 유저 383 이면 0.365937, 
* 유저 383 아니라면 나머지(유저 10)는 0.362127
설명은 하단의  [조건문 단축](#조건문-단축), [map,apply,mapapply](#map,apply,mapapply)참고



```python
temp['weight']= temp.apply(
    lambda x:x['rating']*0.414398 if x['user id']==6 else 
   x['rating']* 0.365937 if  x['user id']==383 else
   x['rating']* 0.362127, axis=1)
```

```python
temp
```

내림차순 정렬하니 동점이 많음.

```python
temp.sort_values('weight',ascending=False).iloc[:10]
```

각 영화에 대해서 가중치 가장 높은 5개 

```python
t=temp.groupby('item id').mean().sort_values('weight',ascending=False).iloc[:5]
t
```
