#### Mode
1. Edit mode (입력 모드) 

   -해당 셀 경계 색상 파랑
            - 상단에 아이콘 사라짐
            - 타이핑 하게되면 현재 활성화된 셀(선택된 셀)에 바로 입력됨
            - 모든 전환 방법 : 셀 내부 마우스 오른쪽 클릭 /  `[단축키: Enter/ return ]`
            
            
2. Command mode (명령 모드)   

          - 해당 셀 경계 색상 초록
          - 상단에 연필 아이콘 
          - Command Mode는 셀 경계에서 단축키를 사용할 수 있다.  
          -  모든 전환 방법 : 근처 셀 바깥의 여백에  마우스 오른쪽 클릭 / `[단축키: Esc]`
