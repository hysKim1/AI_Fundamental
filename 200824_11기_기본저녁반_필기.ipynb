### DNN vs CNN 

#### DNN: 위치 기반
- 이미지 영역을 찾은후 가운데로 정렬하여 분류 학습(특정값이 어느 위치에 있냐에 따라서 구분하기 때문)
    - 예.mnist 숫자데이터: 중앙 정렬,동일 크기, 흑백 처리 되어 있음 
- 위치에따라 민감, 모든 특징들이 연산 되어짐.
- 특징을 가장 잘 찾는(loss function  최소화) kernel,bias찾음
- 2 차원적 특성을 알 수 없음 (다른 형태의 사진은 이해를 못함)
    - 고양이가 정면으로 앉아 있는 자세만 인식 가능하나 숨어있거나 누워있는 등의 다른 형태를 인식 못함



#### CNN: 특성 기반
- Yann LeCun: Convoution을 통해서 특징을 학습을 통해서 발견
    - photoshop의 custom filter : 이미지에 특성을 바꿔줌
    - 이미지 x 입력한 행렬 = > 특성 파악/변화
        - blur, edge추출
- windowing :특징을 알고 있는 채로 이미지에서 쭉 찾아다니다가 눈,코,귀,입 찾음
- 정사각 형태로 특징을 추출하여 2차원적인 공간성도 가져감

하단의 이미지는 custom filter예


### 합성곱
- sliding했을 때 해당 각각 동일한 위치별로 곱하여 합침 
- 특징 잇는 갯수 만큼 전체이미지를 훑어가면서 가장 특성에 합을 더해서 가장 큰 값을 갖는 것으로 특징 판담함

예1. 특성이 동일한 경우 큰값을 갖음


예2. 특성이 동일하지 않은경우 0에 가까운 값




### padding
- 위와 같이 합성곱을 하게 되면중심의 값들이 여러번 연산하게 되면서 특징을 더 학습하 고 가장자리는 덜 학습이 되는 경우가 있음 


zero padding - 나머지를 0 으로 채워서 가장자리를 잘 파악하게 되어 과적합 방지함
- 0으로 채우는 이유는 계산에서 영향력을 없애기 위함



### Feature Map




- 실제 convlolution  곱학고 합한 결과를 Feature Map으로 나옴
    - 이미지 20x20
    - convolution3x3
    - 결과 Featuremap 18x18 : 곱,합을 하나의 값으로 얻음
        - 특징을 대표하는것을 찾을 때 사용하는 기법
        
-  **feature map 기반으로 변환된 이미지로 학습**을 시킴
- 고양이 특징이 어디서 나타날지 모르지만 feature map의  convolution의 값이 큰 경우가 특징이 있는 경우
- 이미지의 크기와 상관없이 특징 파악이 가능해지면서 성능 향상

[CNN 이용한 MNIST 예시 keras.js](https://transcranial.github.io/keras-js/#/mnist-cnn)
 - 학습한 결과를 시각적으로 볼수 있도록 제공
 
- Activation :레이어가 많아질수록 새로운 공간을 왜곡해서 비선형데이터를 선형으로 분류 가능해짐
- convolution할수록 추상적인 특징이 곱해진 결과값을 갖음
- 한번더 하면 
- 특징이 곱해진 결과 

3x3 32가지 특징이미 만들어져 있음


- 특성 이미지가 있는데 가장 잘 구분하는 더 추상적인 특성을 학습을 통해서 찾음
- 학습은 동일 : 원래 값에 곱해서 새로운 이미지를 특징과 곱해져 변화된 이미지를 1차원으로 변화시켜서 기계학습 사용해서 분류하는 기법


각각의 위치에 서로 곱하여 더하여 스칼라 값
진행 순서 좌-우, 상-하로 이동



#### stride 
- 몇 칸씩 움직이는지

####  kernel

#### pooling

하단의 이미지는 2 strides(2칸씩 이동)하여 4x4 -> 2x2


#### 학습

### MaxPool2D
- 합성곱한 결과를 최대값만 pooling을 통해 크기를 줄이는 기법
- 크기는 줄어들지만 고유의 최대값을 통해서 특성을 갖고 있음

```python
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(tf.keras.layers.Conv2D(32, (3, 3), activation='relu'))
model.add(tf.keras.layers.MaxPool2D())
```

```python
model.summary()
```

이미지가 점점 더 단순해짐

```python
p1=model.predict(X_train[0][np.newaxis])
plt.imshow(p1[0][:,:,2], cmap='binary')  #32개 이미지중 1번째 선ㅔ
```

최종 모델

dense,perceptron 정한 기준에는  마법 같은 공식 없음 
- 학습을 통해서 loss최소화

```python
model=tf.keras.models.Sequential()
 #convolution & pooling  역할 알아햐함
model.add(tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(tf.keras.layers.MaxPool2D(2))
model.add(tf.keras.layers.Conv2D(32, (3, 3), activation='relu'))
model.add(tf.keras.layers.MaxPool2D(2))
### 위에 모델을 사용하면 특성값 이용해서 학습함
model.add(tf.keras.layers.Flatten())  #예측 모델 생성 
model.add(tf.keras.layers.Dense(64,activation='relu'))
model.add(tf.keras.layers.Dense(10,activation='softmax'))  #10개 예측
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=1, batch_size=1000) #학습데이터가 많아 오래걸림
```
