### groupby
- 이전 시간에 groupby가 split-apply-combine 한다고 가볍게 언급

#### Split-Apply-Combine Strategy 

- Hadley Wickham
    - R에서 대부분의 패키지 만듦
    - Pandas는 R의 영향을 많이 받음
  
  
#### 1. split 
    - 데이터를 기준에 따라 그룹으로 나눔
#### 2. apply 
    - 각 그룹에 독립적으로 함수 적용
#### 3. combine:
    - 데이터구조에 위의 결과를 함침
참고 : Pandas 공식문서 : https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html


참고 : H.Wickham, "Split-Apply-Combine Strategy for Data Analysis"

```python
tips= sns.load_dataset('tips') 
tips.groupby('sex')    #split
```

```python
tips.groupby('sex').mean() #apply & combine(자동)
```

groupby( index=False)

```python
tips.groupby('sex',as_index=False).mean()
```

### pivot_table
- excel의 pivot table 과 비슷
- 사용자가 간편하게 열, 인덱스,값 지정하여 교차표 형식으로 표현
    - values: numpy에서 데이터 값

- `pivot`: index 에 중복이있으면 불가하여 이것의 일반화인 `pivot_table`은 중복도 가능

```python
tips.pivot_table(index='sex',columns=['day','time'],values='tip',aggfunc='max', margins=True)
```

### crosstab
- 2개이상의 인자로 교차표 생성
- pd.crosstab 은 메소드 방식이 없고 function 방식
- 디폴트로 각 행,열 별 갯수(빈도수,frequency)를 반환
    - `pd.crosstab(index= , columns=)`
- 비율  기반 교차표
    - `pd.crosstab(index= , columns=,noarmalize=true)`
- 집계 함수 기반 교차표
    - `pd.crosstab(index= , columns=,values= ,aggfunc= )`

```python
pd.crosstab(index=tips.sex, columns=tips.day) #빈도수
```

```python
pd.crosstab(index=tips.sex, columns=tips.day,normalize=True) #비율
```

집계함수 사용시에는 값 지정해야함

```python
pd.crosstab(index=tips.sex, columns=tips.day,values=tips.tip,aggfunc='mean')
```

## stack vs unstack
- pivot_table 과 밀접한 관련
- 이 메소드는 multiindex에 사용하려 고안됨
    - 인덱스가 여러개이므로 레벨 선택 가능
    - stack: columns -> index
    - unstack: ides -> columns
    
    
    
(출처: Pandas공식문서https://pandas.pydata.org/docs/user_guide/reshaping.html    )
    
### stack 
- long format
##### SCI
stack은 column 에 있는내용을 index로 만든다

```python
tips.groupby(['sex','day']).mean() #원래 데이터
```

total_bill,tip, size 열들이 인덱스로 됨

```python
tips.groupby(['sex','day']).mean().stack() #
```

```python
tips.groupby(['sex','day']).mean().stack(0)
```

### unstack
- index를 column 으로 옮김
    - wide format
- 아래 예시에서는 'day'
- `unstack() `
    - index 중 가장 마지막 레벨의 인덱스를 인덱스에서 열

```python
tips.groupby(['sex','day']).mean()
```

```python
tips.groupby(['sex','day']).mean().unstack()
```

```python
tips.groupby(['sex','day']).mean().unstack('day')
```

```python
tips.groupby(['sex','day']).mean().unstack(1) #default
```

```python
tips.groupby(['sex','day']).mean().unstack(0)
```
