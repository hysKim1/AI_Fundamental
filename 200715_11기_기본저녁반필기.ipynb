## Numpy( Numerical Python)


- **벡터, 행렬 연산(arrray programming)을 위한 연산하는 python library**
- **빠른 연산과 편리성**
    - C로 구현
    - 연산,처리 속도도 빠름
    - **Vectorization : CPU 갯수만큼 동시에 처리**할수 있도록 지원
        -  계산이 클수록 리소스 소요됨
    - **빠른 속도와 편이**
        - broadcasting(자동으로 형태 변환)
- 많은 과학 계산 라이브러리가 Numpy 기반
- 모든 데이터 구조가 **벡터/행렬연산** 기준

- Glue Language
    - 풀(glue) 처럼 다른 언어와 잘 붙기때문에 붙여짐
        -   numpy는 파이썬으로 만들게 아닌 C로 만듦
        - 빠른 성능이 필요한경우 C와 같은 언어 이용
        - 예. Cython
- 막강한 영향력
    - 파이썬이 Numpy 연산자도 만들어줄 정도

### array

### eye 
- 삼각행렬 만들때 유용 
- `np.eye(n)`
    * n: 행의수, m 열의수
    * k: 대각원소 위치
        * k=0 : default로 가운데 대각선에 위치
        * k>0 : upper diagonal
        * k<0 : lower diagonal
    

```python
z=np.eye(3)#k=0
z
```

```python
z= np.eye(3,k=1)
z
```

```python
z= np.eye(3,k=2)
z
```

```python
z= np.eye(3,k=-1)
z
```

### Mutable or Immutable 
- mutable &immutable의 중간
    *  functional paradigm 에서 mutable 안씀
    * 쉬운 사용
- 다른것을 기반으로한 범용성(먼저 배운이유)
#### tensor 
- numpy 확장판
- **immutable** 
- GPU 지원




| List  | Array |
|:------|:------|
| linked list 로 데이터가 구조화 됨 <br> 계속해서 다시 참조하여 찾아야함 | 메모리에 연속적으로 만들 수 있음 <br>메모리 공간 비워있을때 후반부터 채워짐|
