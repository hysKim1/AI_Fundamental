`super`:자식 클래스가 부모 클래스로부터 상속받은 멤버를 참조할 때 사용하는 참조 변수


    tf 공식문서 맞춤 레이어 참고: https://www.tensorflow.org/tutorials/customization/custom_layers?hl=ko
    
    
- tf.keras.Layer 클래스를 상속하고 하단처럼 구현

```python
class MyDenseLayer(tf.keras.layers.Layer):
  def __init__(self, num_outputs):
    super().__init__()
    print('__init__')
    self.num_outputs = num_outputs

  def build(self, input_shape):
    print('build')
    print(input_shape)
    self.kernel = self.add_weight("kernel",
                                    shape=[int(input_shape[-1]),
                                           self.num_outputs])
    print(self.kernel.shape)

    
  def call(self, input):
    print('call')
    return tf.matmul(input, self.kernel)
```

4개의 output이며 아직 총 몇개의 데이터가 입력 받을지 모름
 ( None,4)

#### build 없는 경우
- 가능 하긴 하나 기본적으로 `Layer`, `Model` 클래스에서는 대부분 있으므로 상속해서 overriding 하려면 알아야함

```python
class MyDenseLayer(tf.keras.layers.Layer):
  def __init__(self, num_outputs):
    super().__init__()
    print('__init__')
    self.num_outputs = num_outputs

  def call(self, input_):
    print('call')
    self.kernel = self.add_weight("kernel",
                                    shape=[int(input_.shape[-1]),
                                           self.num_outputs])
    print(self.kernel)
    return tf.matmul(input_, self.kernel)
```

#### Model 상속
- 상황에 따라 동적 프로그래밍 가능하여 복잡한 모델 생성 가능
- `Model` class 상속하는 고급 기법
    - \_\_init__  
        - flatten & dense 레이어 2개 
    - call
  
전문가를 위한 빠른 시작 공식 문서 참고:https://www.tensorflow.org/tutorials/quickstart/advanced?hl=ko

```python
#https://www.tensorflow.org/tutorials/quickstart/advanced?hl=ko

class MyModel(tf.keras.models.Model):
  def __init__(self):
    super(MyModel, self).__init__()
    self.flatten = tf.keras.layers.Flatten()
    self.d1 = tf.keras.layers.Dense(128, activation='relu')
    self.d2 = tf.keras.layers.Dense(10, activation='softmax')

  def call(self, x):
    if x.shape[-1]>3 :                      #조건문에 따른 동적 프로그래밍 가능!
        x = self.flatten(x)
        x = self.d1(x)
        return self.d2(x)
    else:
        return self.d2(x)
```

```python
model=MyModel() #안스턴스화
```

```python
dir(model)     # __call__ ->callble->f
```

```python
model(tf.ones(10,5))  # _call_ -> build(),call()
```

```python
#한 가지 예시로 
inputs = tf.keras.Input(shape=(28,28))
x = tf.keras.layers.Dense(10)(inputs)
outputs = tf.keras.layers.Dense(1)(x)
modell = tf.keras.Model(inputs, outputs)

# Activity regularization- 종속성이 다른 경우

modell.add_loss(tf.abs(tf.reduce_mean(x)))
```

```python
modell.compile(loss='sparse_categorical_crossentropy', 
              optimizer='adam')#- fit 사용하기 위해서 compile 해야함`
```

```python
modell.fit(x_train,y_train,epochs=5)  #activity regularization 으로 가능
```

- 보통 `compile (loss, metrics)`에서 제공되지 않는 경우
- 활성화만 요구할 수 있으며(이 경우에는 대상이 없음), 이 활성화는 모델 출력이 아닐 수 있다.

 - ActivityRegularizationLayer 경우에는 activation of a layer만 필요함(target 불필요) 
 - 커스텀한 layer의 call method 로 self.add_loss(loss_value)를 호출 가능.
- 이러한 방식으로 추가된 손실을 학습 동안에 `compile`로 보내어지는 loss에 추가된다.

### callbacks
- 한 번 실행하고 나서 epoch 실행되게 함
- 학습과 검증 과정에서 적용할 keras.callbacks.Callback 인스턴스의 리스트.
    -  학습 과정의 특정 단계에서 적용할 함수의 세트
        - keras.callbacks.Callback :확작해서 커스텀 콜백 생성 가능
        - keras.callbacks.BaseLogger
        - keras.callbacks.TerminateOnNaN
        - keras.callbacks.ProgbarLogger
        - keras.callbacks.History
        - keras.callbacks.ModelCheckpoint
        - keras.callbacks.RemoteMonitor
        - LearningRateScheduler
        - keras.callbacks.TensorBoard
        - keras.callbacks.ReduceLROnPlateau
        - keras.callbacks.CSVLogger

keras 공식 문서 참고 : https://keras.io/ko/callbacks/#_1


```python
# 각 batch에 대한 loss  내역 기록
class LossHistory(keras.callbacks.Callback): #상속 -> self.model  접근 가능
    def on_train_begin(self, logs={}):
        self.losses = []

    def on_batch_end(self, batch, logs={}):
        self.losses.append(logs.get('loss'))

        
model = Sequential()
model.add(Dense(10, input_dim=784, kernel_initializer='uniform'))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy', optimizer='rmsprop')

history = LossHistory()  #인스턴스화 한 후 fit에 사용
model.fit(x_train, y_train, batch_size=128, epochs=20, verbose=0, callbacks=[history])

print(history.losses)
```

```python
# epoch마다 모델 저장
from keras.callbacks import ModelCheckpoint

model = Sequential()
model.add(Dense(10, input_dim=784, kernel_initializer='uniform'))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy', optimizer='rmsprop')

'''
만약 검증에 대하 loss가 감소하다면 각 epoch  후의 모델의 가중치를 저장함
'''
checkpointer = ModelCheckpoint(filepath='/tmp/weights.hdf5', verbose=1, save_best_only=True)
model.fit(x_train, y_train, batch_size=128, epochs=20, verbose=0, validation_data=(X_test, Y_test), callbacks=[checkpointer])
```

### trainable

- 'freezing' 이라 불리우며 `fit`학습 시킬때 weights 업데이트 시키지 않고 그대로 가져오기 때문

참고: https://www.tensorflow.org/guide/keras/transfer_learning

```python
inputs=tf.keras.Input(shape=(28,28))
flatten=tf.keras.layers.Flatten()(inputs)
layer1=tf.keras.layers.Dense(128,activation='relu')(flatten)

layer1.trainable=False                            # 추가 frozen layer1

dropout=  tf.keras.layers.Dropout(0.2)(layer1)
output= tf.keras.layers.Dense(10, activation='softmax')(dropout)
model1=tf.keras.models.Model(inputs=inputs,outputs=output)


model1.compile(loss='sparse_categorical_crossentropy')
```

## 자동 미분
- 학습: gradient descent 통해  loss function 최소화하는 파라미터 찾음
    - loss function : 입력 & 실체값 차이를 함수로 표현
    - gradient descent : 현재 위치와 learning rate곱하여 이동하면서 loss/cost function 최소값(미분값=0)을 주는 파라미터를 찾음
        - gradient : 특정점의 기울기
        - learning rate $\alpha$: 현재 그래디언트에서 다음 위치까지 얼마만큼 가야하는지
        - 미분 : 하나의 변수      
        - 편미분: 변수가 여러개 인 경우 각각에 대해 미분)
            - 자동 미분 :auto differentiation
                - tensorflow  지원 4가지: `train`,`fit`, `train_on_batch`, `GradientTape`

## 4.GradientTape
- 1970년대에 인기 저장 매체로 tape 사용하여 영상물 기록
- 주어진 입력에 대한 연산의 gradient에서 최소점을 찾기위해 학습한 것이 어떻겨 변화하는지 편미분값 저장
- 가장 저주순
- 미분하기위해서 미분하는 함수를 GradientTape에 넣으면 자동미분 할 수 있도록 구조체로 변경
    - 변수가 여러개 이므로 편미분을 통해서함
