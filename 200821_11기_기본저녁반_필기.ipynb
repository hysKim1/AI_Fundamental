### Gradient Descent Algorithms
 - loss의 최소값을 찾아가는 과정의 다양한 방식
 -  가중치$w_{new} = w -\alpha \frac{∂L}{∂w}$
    - #### step 방향 $\frac{∂L}{∂w}$ :step 방향
        - 접선: loss function을 weight으로 미분
         - 기울기 +,- 
    - #### step size
        -  learning rate $ \alpha $ : step size(보폭 크기)
        - 값이 작으면 찾아가는 과정이 너무 느려짐
        - 값이 크면 최소값을 지나쳐 갈  수 있음


GD(Graidnet Descent):
- 전체 데이터 만큼  gradient descent 함

SGD(Stochastic Graidnet Descent):
- batch size 만큼 랜덤하게 추출하여 gradient descent 함
- 더 빨리,자주 업데이트

Momentum
- 기울기 바향이 다양한데 동일한 뱡향 일때 가중치를 두어 더 크게 이동 (왔다 갔다 하게 됨)
- 관성,탄력, 가속도
- GD에 탄성을 줌

NAG(Nestrov Accelerated)
- Momentum이 이동시킬 방향으로 미리 이동, graident계산
- 불필요하게 왔다갔다 한는것을 줄임(정확도 개선)

Adagrad(Adaptive Gradient Descent)
- 적응형으로 GD를 함
- 기울기 크면 그게 점프 ,작으면 작게 점프
- 변수의 업데이트가 잦으면 학습률을 줄여 보폭도 줄임

AdaDelta(Adaptive Delta)
- 0에 가까우면 잘 못찾아서 멈추는 현상을 막음

RMSProp
- Adagrad의 보폭 민감도를 보완하여 이전 상황을 고려
- 기본으로 사용

Adam
- RMSProp + Momentum
- **step 방향, size 모두 고려**하여 가중치로 적절히 조정
- 기본으로 사용

NAdam
- RMSProp + NAG


속도도 빠르고 정확도도 좋은 새로운 알고리즘들이 계속해서 나옴(점점 보완)

1. 모델 정의
    - $ f(x) = x * W + b $ 선형 모델

```python
class Model(object):
  def __init__(self):
    # 변수를 (5.0, 1.0)으로 초기화하여 찾아나가도록 할 예정
    self.W = tf.Variable(5.0)
    self.b = tf.Variable(1.0)

  def __call__(self, x):
    return self.W * x + self.b  #선형 함수
```

```python
model = Model()

model(tf.ones((3,4)))
```

선형 함수 : 5. x 1 + 1. = 6.

2. 손실 함수 정의
    loss
    - 모델의 예측이 실제값과 얼마나 잘 일치하는지를 측정
    - MSE(평균제곱오차) :실제값 , 예측값간의 차이 제곱의 합의 평균
        - $\sum \frac{(y_i -\hat y_i)^2}{N} $

- loss  최소화 하는것은 ML마다 다름 
    - 신경망 모델에서는 graident descent 방식 이용
    - 최소값을 찾아주는 w,b 찾아야함

```python
def loss(predicted_y, desired_y):
  return tf.reduce_mean(tf.square(predicted_y - desired_y))
```

3. 훈련 데이터 가져오기

훈련 데이터
- 정규 분포를 따르는 1000개를 random하게 만듦

가짜 데이터 
- 정규 분포를 따르는 1000개를 random 하게 만듦 
- 실제와 어긋나게 만들기 위해서 랜덤하게 생성한 noise를 훈련데이터 input에 더함

```python
TRUE_W = 3.0
TRUE_b = 2.0
NUM_EXAMPLES = 1000

inputs  = tf.random.normal(shape=[NUM_EXAMPLES])  #랜덤하게 생성

noise   = tf.random.normal(shape=[NUM_EXAMPLES]) # 편차 추가하여 가짜데이터 생성
outputs = inputs * TRUE_W + TRUE_b + noise       #선형 모형 
```

```python
model(inputs)
```

- 초기화 된 모델  시각화
    - 데이터(파랑색 점)을 가장 잘 표현하는 모델(붉은 점)을 찾는것
    - 해당 모델과 noise통해 만든 데이터간의 차이를 최소화 하는 w,b 찾기

```python
import matplotlib.pyplot as plt  

plt.scatter(inputs, outputs, c='b')
plt.scatter(inputs, model(inputs), c='r')
plt.show()

print('현재 손실: '),
print(loss(model(inputs), outputs).numpy())  #초기값 
```

`GradientTape`: 미분값을 다 기록함

위의 데이터에서 하단과 같이 미분하여 w,b 값 업데이트 하면 다음과 같이 더 가까운 값으로 예측함

4. 훈련 데이터에서 실행, 데이터에 최적화하기 위해 "옵티마이저(optimizer)"를 사용한 변수 조정
- gradient descent 이용 w,b 업데이트 하면서  손실 감소
- 훈련 루프 

```python
#미분하고   입력
with tf.GradientTape() as t:
    current_loss = loss(model(inputs), outputs) #loss function
    
ww,bb= t.gradient(current_loss, [model.W, model.b])#weight,loss로 gradient
ww,bb # weight, bias 변경됨
learning_rate=0.1
#한번 사용하고 사라지기 때문에 with 에 사용
model.W.assign_sub(ww*learning_rate)  # 계산하여 업데이트한 값으로 재할당하여 변경 
model.b.assign_sub(bb*learning_rate)
```

#### updating weights

- assign_sub
- 가중치$w_{new} = w -learning rate \frac{∂L}{∂w}$
    - model.W -= ww(learning_rate) ->  model.W - ww x learning rate 

할당 방법
- a=1
- a=b=1
- a,b=1,2
- a,b*=1,2,3
- a +=1      ->assign_sub 이 여기 해당

여러번 실행하면서 점점 더 붉은점이 푸른색 점을 잘 대표하게 됨(loss 최소화)

```python

with tf.GradientTape() as t:
    current_loss = loss(model(inputs), outputs) #loss function
    learning_rate=0.1
    ww,bb= t.gradient(current_loss, [model.W, model.b])#weight,loss로 gradient
    model.W.assign_sub(ww*learning_rate)  # 계산하여 업데이트한 값으로 재할당하여 변경 
    model.b.assign_sub(bb*learning_rate)

    plt.scatter(inputs, outputs, c='b')
    plt.scatter(inputs, model(inputs), c='r')
    plt.show()

    print('현재 손실: ',loss(model(inputs), outputs).numpy())  #loss 줄어들음
```

이 데이터를 잘 대표하는 모델을 찾기위해서 위의 과정을 반복함 
- 하단에서 반목분 이용하여 데이터를 가장 잘표현하는 $ y= \alpha x+\beta $ 에서 $ \alpha,\beta $찾음

```python
for i in range(10):
    with tf.GradientTape() as t:
        current_loss = loss(model(inputs), outputs) #loss function

    ww,bb= t.gradient(current_loss, [model.W, model.b])

    model.W.assign_sub(ww*learning_rate)  # 계산하여 업데이트한 값으로 할당 
    model.b.assign_sub(bb*learning_rate)
                
    plt.scatter(inputs, outputs, c='b')
    plt.scatter(inputs, model(inputs), c='r')
    plt.show()

    print('현재 손실: ',loss(model(inputs), outputs).numpy()) #손실함수가 점점 작아지면서 데이터를 잘 나타냄
```

w, b변화하는 과정 시각화
- epoch 늘어날수록 초기화해 놓은 w=5,b=1에서  실제 w=3,b=2에 가까워 지면서 손실이 급격히 줄어듦

```python
def train(model, inputs, outputs, learning_rate):
  with tf.GradientTape() as t:
    current_loss = loss(model(inputs), outputs)
  dW, db = t.gradient(current_loss, [model.W, model.b])
  model.W.assign_sub(learning_rate * dW)
  model.b.assign_sub(learning_rate * db)


model = Model()

# 도식화를 위해 W값과 b값의 변화를 저장합니다.
Ws, bs = [], []
epochs = range(10)
for epoch in epochs:
  Ws.append(model.W.numpy())
  bs.append(model.b.numpy())
  current_loss = loss(model(inputs), outputs)

  train(model, inputs, outputs, learning_rate=0.1)
  print('에포크 %2d: W=%1.2f b=%1.2f, 손실=%2.5f' %
        (epoch, Ws[-1], bs[-1], current_loss))

# 저장된 값들을 도식화합니다.
plt.plot(epochs, Ws, 'r',
         epochs, bs, 'b')
plt.plot([TRUE_W] * len(epochs), 'r--',
         [TRUE_b] * len(epochs), 'b--')
plt.legend(['W', 'b', 'true W', 'true_b'])
plt.show()
```

```python
w = tf.Variable(tf.random.normal((3, 2)), name='w')
b = tf.Variable(tf.zeros(2, dtype=tf.float32), name='b')
x = [[1., 2., 3.]]

with tf.GradientTape(persistent=True) as tape:
  y = x @ w + b
  loss = tf.reduce_mean(y**2)
```

```python
with tf.GradientTape() as tape:
  # Forward pass - 실제 레이어에 여러 값을 집어넣음
  y = layer(x) 
  loss = tf.reduce_mean(y**2) #함수

# Calculate gradients with respect to every trainable variable
grad = tape.gradient(loss, layer.trainable_variables)
```

### watch_accessed_variables
- 동시에 연관된것 볼지 안볼지 정함
- 모든 tf.Variables 을 볼지 안 볼지 정함
    - default: `watch_accessed_variables = True'
        - True :해당 tape에서 계산된 모든 값을 추적
        - False : Variable,과 watch로 명시되어 있는것만 추적 

```python
#두 변수 x0,x1 사용했지만  x1 만 y1,y,s랑 연결 되어  gradient 에서도 x1만 연결됨
x0 = tf.Variable(0.0)
x1 = tf.Variable(10.0)

with tf.GradientTape(watch_accessed_variables=False) as tape:
  tape.watch(x1)
  y0 = tf.math.sin(x0)
  y1 = tf.nn.softplus(x1)
  y = y0 + y1
  ys = tf.reduce_sum(y)
```

```python
# dy = 2x * dx
grad = tape.gradient(ys, {'x0': x0, 'x1': x1})

print('dy/dx0:', grad['x0'])
print('dy/dx1:', grad['x1'].numpy())
```

```python
#두 변수 x0,x1 사용했지만  x1 만 y1,y,s랑 연결 되어  gradient 에서도 x1만 연결되었지만 다 나옴 
x0 = tf.Variable(0.0)
x1 = tf.Variable(10.0)

with tf.GradientTape(watch_accessed_variables=True) as tape:
  tape.watch(x1)
  y0 = tf.math.sin(x0)
  y1 = tf.nn.softplus(x1)
  y = y0 + y1
  ys = tf.reduce_sum(y)
```

```python
# dy = 2x * dx
grad = tape.gradient(ys, {'x0': x0, 'x1': x1})

print('dy/dx0:', grad['x0'])  #true 인경우 나오긴 함
print('dy/dx1:', grad['x1'].numpy())
```

watch_accessed_variables
- 기본으로 모든 변수를 추적하도록 됨
- 추적안하도록 변경하면 gradient와 연결된 값만 나옴
- GradientTape이 활성화 되는 동안에 trainable variables과 watch로 명시한값들을 자동적으로 추적할지 안할지 제어

```python
x=tf.Variable([1.,2.,3.]) #상수값은 안나오므로 watch 이용해서 변수처럼 사용 가능
with tf.GradientTape(watch_accessed_variables=False, persistent=True) as ttt:
    y=x**2
    z=y**2
ttt.gradient(y,x) # 안나옴
```
